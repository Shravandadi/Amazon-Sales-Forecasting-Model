## Amazon Sales Dataset EDA Documentation

This document provides a step-by-step explanation of the Exploratory Data Analysis (EDA) performed on the Amazon Sales Dataset.

### 1. Data Loading and Inspection

*   **Code:** `data = pd.read_csv('/content/amazon_dataset.csv')\ndata.head()`
*   **Explanation:** The first step is to load the dataset into a pandas DataFrame. The `pd.read_csv()` function is used to read the data from the specified file path. `data.head()` displays the first five rows of the DataFrame, allowing for a quick initial inspection of the data and its structure.

*   **Code:** `data['user_id'] = data['user_id'].astype(str).str.split(',')\ndata['user_name'] = data['user_name'].astype(str).str.split(',')`
*   **Explanation:** These lines aim to separate multiple user IDs and names within a single cell, assuming they are comma-separated. `astype(str)` is used to ensure the column is treated as strings, handling potential non-string entries like `NaN`. `str.split(',')` then splits the strings by the comma delimiter, resulting in lists of strings in each cell.

*   **Code:** `data`
*   **Explanation:** This code simply displays the entire DataFrame, allowing you to view all rows and columns of the loaded data.

### 2. Data Cleaning

*   **Code:** `df = data.dropna()\ndf.head()`
*   **Explanation:** This code attempts to clean the dataset by removing rows with any missing (NaN) values. `data.dropna()` creates a new DataFrame `df` with the rows containing missing values removed. `df.head()` displays the first five rows of this new DataFrame. **However, as observed in the output, this resulted in an empty DataFrame because the 'user_id' and 'user_name' columns had missing values in almost all rows.**

*   **Code:** `data.isnull()`
*   **Explanation:** This code checks for missing values in the original `data` DataFrame. It returns a DataFrame of boolean values, where `True` indicates a missing value and `False` indicates a non-missing value. This helps in identifying which columns and rows have missing data.

### 3. Data Manipulation: Filtering, Sorting, and Grouping

*   **Code:** `df.info()`
*   **Explanation:** This provides a summary of the `df` DataFrame, including the data types of each column and the number of non-null values. Since `df` is empty after dropping rows with missing values, the output shows that there are 0 non-null entries for all columns.

*   **Code:** `data.describe()`
*   **Explanation:** This code generates descriptive statistics for the numerical columns in the `data` DataFrame. The output provides insights into the central tendency, dispersion, and shape of the data's distribution. For example, it shows the count, mean, standard deviation, minimum, maximum, and quartile values for 'discounted\_price', 'discount\_percentage', 'user\_id', and 'user\_name'. Note that the statistics for 'user\_id' and 'user\_name' are not meaningful in this context as they were converted to lists.

*   **Code:** `data.sort_values(by='rating', ascending=True)`
*   **Explanation:** This code sorts the `data` DataFrame based on the values in the 'rating' column in ascending order. This allows for easy inspection of products with the lowest ratings.

*   **Code:** `data.groupby('rating')`
*   **Explanation:** This code groups the `data` DataFrame by the unique values in the 'rating' column. This operation is often a precursor to performing aggregations or analyses on subsets of data based on their rating. The output is a `GroupBy` object, which can be used for further operations.

*   **Code:** `grouped_data = data.groupby('product_id')\ngrouped_data`
*   **Explanation:** This code groups the `data` DataFrame by the 'product\_id' column. This is useful for analyzing data on a per-product basis. The output is a `GroupBy` object.

### 4. Mathematical Operations

*   **Code:** `data['discounted_price'] = data['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)\naverage_discounted_price = data['discounted_price'].mean()\nprint(f"The average discounted price is: {average_discounted_price}")`
*   **Explanation:** This code first cleans the 'discounted\_price' column by removing the currency symbol ('₹') and commas, and then converts the column to a float data type. This is necessary to perform numerical calculations. It then calculates the mean (average) of the 'discounted\_price' column and prints the result.

*   **Code:** `mode_discounted_price = data['discounted_price'].mode()\nprint(f"The mode of discounted price is: {mode_discounted_price}")`
*   **Explanation:** This code calculates the mode (the most frequently occurring value) of the 'discounted\_price' column and prints it.

*   **Code:** `data[data['discounted_price'] == 199.0]`
*   **Explanation:** This code filters the `data` DataFrame to display only the rows where the 'discounted\_price' is exactly 199.0. This allows for inspection of products with this specific discounted price.

*   **Code:** `sum_discounted_price = data['discounted_price'].sum()\nprint(f"The sum of discounted price is: {sum_discounted_price}")`
*   **Explanation:** This code calculates the sum of all values in the 'discounted\_price' column of the `data` DataFrame and prints the result.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (replace with your actual data)
actual_prices = data['actual_price'].head(10)  # Taking the first 10 values for example
discounted_prices = data['discounted_price'].head(10)

# Create bar positions
bar_positions = np.arange(len(actual_prices))

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(bar_positions, actual_prices, width=0.4, label='Actual Price')
plt.bar(bar_positions + 0.4, discounted_prices, width=0.4, label='Discounted Price')

# Add labels, title, and legend
plt.xlabel('Product Index')
plt.ylabel('Price')
plt.title('Comparison of Actual and Discounted Prices')
plt.xticks(bar_positions + 0.2, bar_positions)  # Center the x-axis ticks
plt.legend()

# Show the chart
plt.show()
```

A scatter plot or histogram for distributions.

```python
# A scatter plot for distributions of discount_percentage.

import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame with a 'discount_percentage' column
# Convert 'discount_percentage' to numeric, removing '%' and converting to float
data['discount_percentage'] = data['discount_percentage'].str.replace('%', '').astype(float)

plt.figure(figsize=(8, 6))  # Adjust figure size if needed

plt.scatter(data.index, data['discount_percentage'], alpha=0.5)  # alpha for transparency

plt.xlabel('Index')  # Or any other relevant x-axis label
plt.ylabel('Discount Percentage')
plt.title('Distribution of Discount Percentage')
plt.grid(True)  # Add a grid for better readability
plt.show()
```
